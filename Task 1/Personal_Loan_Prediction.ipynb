# Null Value Treatment

```python
data['Gender'].fillna(data['Gender'].mode()[0], inplace=True)
data['Home Ownership'].fillna(data['Home Ownership'].mode()[0], inplace=True)
data['Online'].fillna(data['Online'].mode()[0], inplace=True)
data['Income'].fillna(data['Income'].median(), inplace=True)
```

```python
data.dtypes
```

# Normalization of Income Using Log Transformation  

```python
data['Income_log']=np.log(data['Income'])
data['Income_log'].hist(bins=20)
```

```python
data['Mortgage'] = data['Mortgage'].astype(float)
```

# Normalization of Mortgage Using Min-Max Scaling

```python
min_non_zero = data.loc[data['Mortgage'] > 0, 'Mortgage'].min()

# Replace zeros with the minimum non-zero value
data['Mortgage_normalized'] = data['Mortgage'].replace(0, min_non_zero)

# Perform Min-Max scaling
data['Mortgage_normalized'] = (data['Mortgage_normalized'] - min_non_zero) / (data['Mortgage'].max() - min_non_zero)
```

```python
df = data
```

# Converting categorical variables to numeric

```python
df.loc[df["Gender"] == 'M', "Gender"] = 1
df.loc[df["Gender"] == 'F', "Gender"] = 2
df.loc[df["Gender"] == 'O', "Gender"] = 3
df["Gender"] = df["Gender"].astype(int)
```

```python
df.loc[df["Home Ownership"] == "Home Mortage", "Home Ownership"] = 1
df.loc[df["Home Ownership"] == "Home Owner", "Home Ownership"] = 2
df.loc[df["Home Ownership"] == "Rent", "Home Ownership"] = 3
df["Home Ownership"] = df["Home Ownership"].astype(int)
```

```python
df.columns
```

```python
#Old Income and Mortgage are removed and a new_df is formed
columns_to_drop = ['Income', 'Mortgage']

new_df = df.drop(columns_to_drop, axis=1)
```

# Splitting Dataset

```python
X = new_df.drop('Personal Loan',1)
y = new_df['Personal Loan']
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_cv, y_train, y_cv = train_test_split(X,y, test_size=0.3)
```

# Model Training and Evaluation

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
  
from sklearn import metrics
  
knn = KNeighborsClassifier(n_neighbors=3)
rfc = RandomForestClassifier(n_estimators = 7,
                             criterion = 'entropy',
                             random_state =7)
svc = SVC()
lc = LogisticRegression()
  
# making predictions on the training set
for clf in (rfc, knn, svc,lc):
    clf.fit(x_train, y_train)
    Y_pred = clf.predict(x_train)
    print("Classification_Report of ",
          clf.__class__.__name__,"= \n",
          metrics.classification_report(y_train,
                                     Y_pred))
```

# Predictions on test set:

```python
# making predictions on the testing set
for clf in (rfc, knn, svc,lc):
    clf.fit(x_train, y_train)
    Y_pred = clf.predict(x_cv)
    print("Classification_Report of ",
          clf.__class__.__name__,"= \n",
          metrics.classification_report(y_cv,
                                     Y_pred))
```

```python
rfc.fit(x_train, y_train)
Y_pred = rfc.predict(x_cv)
print(metrics.classification_report(y_cv, Y_pred))
```
